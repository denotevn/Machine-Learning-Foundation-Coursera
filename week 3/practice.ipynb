```python
import turicreate as tc
products = tc.SFrame('/home/tuandinh/Desktop/Machine Learning Foundations Coursera/week3/data/amazon_baby.sframe')
```

```python
products.head()
```

```python
products['word_count'] = tc.text_analytics.count_words(products['review'])
products.head()
```

```python
#ignore all 3* reviews 

products = products[products['rating'] != 3] 
```

```python
#positive sentiment = 4* or 5* reviews
products['sentiment'] = products['rating'] >=4
```

```python
products.head()
```

### Sentiment model

```python
train_data,test_data = products.random_split(.8, seed=0)
```

```python
sentiment_model = tc.logistic_classifier.create(train_data,
                                                     target='sentiment',
                                                     features=['word_count'],
                                                     validation_set=test_data)
```

### Evalue Sentiment model

```python
sentiment_model.evaluate(test_data)
```

### Selected_word_model

```python
import turicreate as tc
products = tc.SFrame('/home/tuandinh/Desktop/Machine Learning Foundations Coursera/week3/data/amazon_baby.sframe')
```

```python
def word_count(word_counts, column_name):
    if(column_name not in word_counts):
        return 0
    return word_counts[column_name]
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
products['word_count'] = tc.text_analytics.count_words(products['review'])
products.head()
```

```python
for word in selected_words:
    products[word] = products['word_count'].apply(lambda x : word_count(x, word))
```

```python
products.head()
```

```python
for word in selected_words:
    print(word + ": " + str(products[word].sum()))
```

```python
products = products[products['rating'] != 3]
```

```python
products['sentiment'] = products['rating'] >= 4
```

### Create model

```python
train_data, test_data = products.random_split(.9, seed=0)
```

```python
model_selected_word = tc.logistic_classifier.create(train_data, target='sentiment',
                                    features=selected_words,
                                    validation_set=test_data)
```

```python
products['predict_sentiment'] = model_selected_word.predict(products, output_type='probability')
products.head()
```

```python
model_selected_word.evaluate(test_data)
```
